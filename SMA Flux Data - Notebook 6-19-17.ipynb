```python
#compile information from each individual flux file into a single data file
frameList = [xl.parse(name) for name in xl.sheet_names[1:]]
structure = pd.concat(frameList)
```

```python
structure["Name"].value_counts() #see which objects are observed in multiple flux observations
```

```python
#stuff file information into a data structure
ScanSum = pd.DataFrame(structure[["Name","#Scans","configuration","file"]]) #remember, single [] pass indices; but need double brackets as it's a list
ScanSum
```

```python
#testing out ways to categorize data and sum; what's the sum of ALL scans taken in compact?
ScanSum['#Scans'][ScanSum['configuration']=='compact'].sum() 
#what's the number of objects observed in each flux file?
ScanSum['file'].value_counts()
#what's the number of times an objects has been observed in a flux file in 2017?
#ScanSum['Name'].value_counts()
#what's the highest number of scans for each object throughout flux observations in 2017?
Summary = ScanSum.groupby('Name')['#Scans'].sum()
Summary.sort_values(ascending=False, inplace=True)
Summary
```
