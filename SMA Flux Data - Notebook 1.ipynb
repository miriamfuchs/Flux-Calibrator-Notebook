```python
frameList = [xl.parse(name) for name in xl.sheet_names[1:]]
structure = pd.concat(frameList)
```

```python
# where structure["Name"]=="0731+713" provides boolean index for where there is overlap of objects b/w observations 
structure[structure["Name"]=="0721+713"] 
```

```python
ScanSum = pd.DataFrame(structure[["Name","#Scans"]]) #remember, single [] pass indices; but need double brackets as it's a list
ScanSum
```

```python
ScanSum["Observations"] = 1 #create column for counting # of scans throughout observations
ScanSum
```

```python
SummaryOfScans = ScanSum.groupby("Name").sum()                          #Summing # of Scans throughout 
SummaryOfScans.sort_values("#Scans",ascending=False, inplace=True)
SummaryOfScans
```
